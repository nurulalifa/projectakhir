```python
import json
import requests
import pandas as pd
```

```python
response = requests.get('https://pluang.com/api/asset/gold/pricing?daysLimit=1825').text
```

```python
response_info = json.loads(response)
```

```python
response_info
```

```python
requests.get('https://pluang.com/api/asset/gold/pricing?daysLimit=1825').json()
```

```python
a = response_info['data']
```

```python
hasil = a['history']
```

```python
hasil
```

```python
dataset = []
for data in hasil:
  dataset.append([data['sell'], data['buy'], data['installment'], data['updated_at']])
```

```python
dataset_df = pd.DataFrame(data=dataset, columns=['Sell', 'Buy', 'Installment', 'Updated_at'])
```

```python
dataset_df.head(5)
```

```python
dataset_df.to_csv('hasildata.csv')
```
